# Decision Tree
Decision Tree는 학습 데이터의 패턴을 파악하여 class를 예측하기 위한 규칙을 만들어내는 알고리즘 입니다. 임의의 데이터를 Deicison Tree에 입력하면 설정된 규칙에 따라 데이터의 class가 할당됩니다. Skelarn에서 제공하는 Decision Tree 모델을 이용하여 실습을 진행해봅시다.

```python
from sklearn import tree
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import mglearn

from matplotlib import pyplot as plt
```

mglearn에서 제공하는 함수를 이용해 임의의 데이터를 생성하여 Decision Tree을 적용해보겠습니다. 데이터가 어떻게 분포되어있는지 확인하고, Decision Tree를 생성하여 도식화해봅시다.

```python
X,y = mglearn.datasets.make_forge()
mglearn.discrete_scatter(X[:,0],X[:,1],y)
plt.show()
```

```python
clf = tree.DecisionTreeClassifier()
clf.fit(X, y)
```

```python
fig = plt.gcf()
fig.set_size_inches(15, 5)
tree.plot_tree(clf, filled=True)
fig.show()
```

유방감 데이터를 불러와서 암 발생 여부를 예측하기 위한 Decision Tree를 설계해보겠습니다.

```python
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer['data'], cancer['target'])
```

```python
print(X_train.shape)
```

```python
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
fig = plt.gcf()
fig.set_size_inches(30, 10)
tree.plot_tree(clf, filled=True)
fig.show()
```

```python
predict = clf.predict(X_test)
print(accuracy_score(y_test, predict))
```
